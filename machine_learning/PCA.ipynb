```python
#using Principle Component Analysis(PCA) to project 2Ddata onto its principal axes
#using PCA to project higher dimensional data to a lower dimensional feature space
#(PCA helps you filter out redundant, linearly correlated variables and reduce the amount of noise in your data)
#PCA stands for Principal Component Analysis.
#It‚Äôs like putting on special glasses üï∂Ô∏è that let you look at your dots from a smarter direction so:

#You can see the spread better

#You can describe the dots using less information (just the important part)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
mean=[2,4]
matrix = [[3,2],[2,2]]
X=np.random.multivariate_normal(mean=mean,cov=matrix,size=200)
plt.scatter(X[:,0], X[:,1], alpha=0.6)
plt.title("2D Multivariate Normal Distribution")
plt.xlabel("X‚ÇÅ")
plt.ylabel("X‚ÇÇ")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
#performing PCA on the dataset
pca =PCA(n_components=2)  #initializing a 2-component PCA model
X_pca= pca.fit_transform(X)
components = pca.components_
components
plt.scatter(X_pca[:,0], X_pca[:,1], alpha=0.6)
plt.title("2D Multivariate Normal Distribution after PCA transform")
plt.xlabel("X‚ÇÅ")
plt.ylabel("X‚ÇÇ")
plt.axis('equal')
plt.grid(True)
plt.show()

```

```python
pca.explained_variance_ratio_
#This gives you a list (or array) of how much of the original data‚Äôs
#  variance is captured by each principal component.
```

```python
projection_pc1 = np.dot(X, components[0])
projection_pc2 = np.dot(X, components[1])
x_pc1 = projection_pc1 * components[0][0]
y_pc1 = projection_pc1 * components[0][1]
x_pc2 = projection_pc2 * components[1][0]
y_pc2 = projection_pc2 * components[1][1]
plt.figure()
plt.scatter(X[:, 0], X[:, 1], label='Original Data', ec='k', s=50, alpha=0.6)

# Plot the projections along PC1 and PC2
plt.scatter(x_pc1, y_pc1, c='r', ec='k', marker='X', s=70, alpha=0.5, label='Projection onto PC 1')
plt.scatter(x_pc2, y_pc2, c='b', ec='k', marker='X', s=70, alpha=0.5, label='Projection onto PC 2')
plt.title('Linearly Correlated Data Projected onto Principal Components', )
plt.xlabel('Feature 1',)
plt.ylabel('Feature 2',)
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
